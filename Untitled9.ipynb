```python
A=randn(25)
```

```python
sns.rugplot(A)
```

```python
plt.hist(A,alpha=0.3)
sns.rugplot(A)
```

```python
sns.rugplot(A)
xmin=A.min()-2
xmax=A.max()+2
x_axis=np.linspace(xmin,xmax,100)
bandwidth=((4*A.std()**5)/(3*len(A)))**0.2
kernel_list=[]
for i in A:
    kernel=stats.norm(i,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    kernel=kernel/kernel.max()
    kernel=kernel*0.4
    plt.plot(x_axis,kernel,color='grey',alpha=0.5)
    plt.ylim(0,1)
```

```python
sumofkde=np.sum(kernel_list,axis=0)
fig=plt.plot(x_axis,sumofkde,color='indianred')
sns.rugplot(A)
```
